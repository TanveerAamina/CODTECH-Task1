```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
from sklearn.datasets import load_iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
```

```python
# Mapping target values to species names
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
# Display the first few rows of the dataset
print("Dataset Preview:\n", df.head())
```

```python
# 1. Inspect the Dataset
print("\nDataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe)
```

```python
# 2. Check for Missing Values
print("\nMissing Values in Each Column:\n", df.isnull().sum())
```

```python
# 3. Data Distribution
# Histograms
df.hist(bins=15, figsize=(12, 8), edgecolor='black')
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Boxplots
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns=['species']))
plt.title("Boxplots of Features")
plt.show()
```

```python
# 4. Pairplot to Explore Relationships
sns.pairplot(df, hue='species', diag_kind='kde', height=2)
plt.suptitle("Pairplot of Features by Species", y=1.02)
plt.show()
```

```python
# 5. Correlation Heatmap
corr_matrix = df.iloc[:, :-1].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 6. Scatterplot of Two Features with Species
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', hue='species', style='species', palette='Set2')
plt.title("Scatterplot of Sepal Length vs Petal Length")
plt.show()
```

```python
# 7. Identifying Outliers with Boxplots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df.iloc[:, :-1], orient='h', palette='Set3')
plt.title("Outlier Detection using Boxplots")
plt.show()
```
